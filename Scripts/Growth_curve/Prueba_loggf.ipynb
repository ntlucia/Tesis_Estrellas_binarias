```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
data_klausTi = pd.read_csv("DataSet/Inputs/Ti2_data_klaus.txt", delimiter = '\t', usecols = ['lambda', 'loggf', 'ewr1'], header = 0)
data_klausFe = pd.read_csv("DataSet/Inputs/Fe1_data_klaus.txt", delimiter = '\t', usecols = ['lambda', 'loggf', 'ewr1'], header = 0)

growth_c_mineTi = pd.read_csv("DataSet/Outputs/Ti 2_growth_curve_use.dat", delimiter = '\t', header = 0)
growth_c_mineFe = pd.read_csv("DataSet/Outputs/Fe 1_growth_curve_use.dat", delimiter = '\t', header = 0)
Atomic_lines = pd.read_csv('DataSet/Inputs/atomic_lines.tsv', delimiter = '\t',   header = None, low_memory=False, keep_default_na= False)
```
